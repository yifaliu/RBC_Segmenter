## 2) Data preparation

```python
import os
from glob import glob
import skimage.io
import numpy as np
import cv2


def get_avg_std_info(config = None):
    """
    calculate the average and standard deviation of RGB images
    """   
    avg = np.zeros(config.NUM_CHANNEL)
    std = np.zeros(config.NUM_CHANNEL)
    num_img = 0
    
    os.chdir(config.IMG_FOLDER)
#     print(config.IMG_FOLDER)
    imgs = glob('*day*.png')
    
    for img in imgs:
        image = skimage.io.imread(img)[:, :, :3]
        
        for j in range(config.NUM_CHANNEL):
            avg[j] = avg[j] + np.mean(image[:, :, j])
            std[j] = std[j] + np.std(image[:, :, j])
        num_img += 1
        
    print("Detected {} images, the average is {}, the standard deviation is {}".format(num_img, avg, std))
    
    return avg,std
```

```python
import skimage
import skimage.io
import skimage.transform
import pandas as pd
import math
import numpy as np
import skimage.morphology
from skimage.morphology import disk

def augmentor(image, label):
    """Do image shape and color augmentation.
    
    Args:
        image and its corresponding mask, the size should be [channel * height * width]
    """
    n_channels = image.shape[0]
    
    # Shape augmentation
    if np.random.uniform(0, 1) < 0.3:
        # Do 70% projection transform
        scale = np.random.uniform(0.8, 1.25)
        aspect_ratio = np.random.uniform(0.8, 1.2)
        rotation = np.random.uniform(-0.2, 0.2)
        translationX = np.random.uniform(-30, 30)
        translationY = np.random.uniform(-30, 30)
        g = np.random.uniform(-0.002, 0.002)
        h = np.random.uniform(-0.002, 0.002)
        matrix = np.array([[math.cos(rotation) * scale * aspect_ratio, -math.sin(rotation), translationX],
                          [math.sin(rotation), math.cos(rotation) * scale / aspect_ratio, translationY],
                          [g, h, 1]])
        tform = skimage.transform.ProjectiveTransform(matrix=matrix)
        image_aug = np.zeros_like(image, dtype=np.float)
        label_aug = np.zeros_like(label, dtype=np.int)
        for ch in range(n_channels):
            image_aug[ch, :, :] = skimage.transform.warp(image[ch, :, :], tform, preserve_range=True)
        label_aug[:, :] = skimage.transform.warp(label[:, :], tform, preserve_range=True)
        # Put label_aug to [0, 1, 255]
        label_aug[np.logical_and(label_aug != 0, label_aug != 1)] = 255
        image = image_aug
        label = label_aug
    if np.random.uniform(0, 1) < 0.6:
        # Do 50% vertical flipping
        image = image[:, ::-1, :]
        label = label[::-1, :]
    if np.random.uniform(0, 1) < 0.6:
        # Do 50% horizontal flipping
        image = image[:, :, ::-1]
        label = label[:, ::-1]

    # Color augmentation
    # 1) add a global shifting for all channels
#     image = image + np.random.randn(1)[0] * 0.3

    # 2) add a shifting & variance for each channel
#     for ch in range(n_channels):
#         image[ch, :, :] = image[ch, :, :] * np.clip(np.random.randn(1)[0] * 0.2 + 1, 0.8, 1.2) + np.random.randn(1)[0] * 0.2
        
    return image, label
```

```python
colorKeyGT = {
    'Non-healing': [255,255,81]
}

def generator(config, ave, std, indicator = "train", index=None, batch_size=None, shuffle=True, augmentation=True):
    """Image generator for training and inference.
    Training mode: randomly yield the images and labels of batch size.

    Args:
        config: The Config object.
        indicator: "train" or "test"
        index: Index of image used for testing ("train" mode: exclude index, "test" mode: exclude others except index)
        batch_size: Number of samples in each batch.
        shuffle: Whether shuffle image indexs.
        augmentation: Whether do image augmentation.
        
    Yields:
        img_batch: [batch_size, n_channels, height, width]
        label_batch: [batch_size,height, width]
    """
    
    
    os.chdir(config.IMG_FOLDER)
    imgs = glob("*day*.png")
   # imgs = [ele.split('_day')[0] + '.jpg' for ele in imgs]  
    num_total = len(imgs)
    imgs_test = [imgs.pop(index)]
    imgs_train = imgs
    
    num_train = len(imgs_train)
    num_test = len(imgs_test)
    
#     print('total image: {}---training:{} -- testing:{}'.format(num_total,num_train,num_test))
    
    
    if num_total<=1:
        print('Warning: the images are not enough to train a deep learning algorithm!')
        
    if indicator == "train":
        np.random.shuffle(imgs_train)
        # Number of steps in each epoch
        if not batch_size:
            batch_size = config.BATCH_SIZE
        n_steps = int(math.ceil(num_train/batch_size))
        img_dataset = imgs_train
        img_size = num_train
    elif indicator == "test":
        batch_size = 1
        n_steps = num_test
        img_dataset = imgs_test
        img_size = num_test
    else:
        print("Error:The indicator should be 'train' or 'test'")
        
    
    for n_step in range(n_steps):
        indexs_imgs = img_dataset[n_step * batch_size:min((n_step + 1) * batch_size,img_size)]
        indexs_masks = [ele.split('_day')[0] + '.jpg' for ele in indexs_imgs]
        
        img_batch = np.zeros((batch_size,config.NUM_CHANNEL,config.IMG_SHAPE[0],config.IMG_SHAPE[1]),dtype=np.float32)
        label_batch = np.zeros((batch_size,config.IMG_SHAPE[0],config.IMG_SHAPE[1]),dtype=np.float32)
        
        for index,img_file in enumerate(indexs_imgs):
            img_file_temp = os.path.join(config.IMG_FOLDER,img_file)
            mask_file_temp = os.path.join(config.IMG_FOLDER,indexs_masks[index])
            
            img_temp = skimage.io.imread(img_file_temp)[:,:,:3]
            mask_temp = skimage.io.imread(mask_file_temp)[:,:,:3]
            
            img_temp = skimage.transform.resize(img_temp, output_shape=config.IMG_SHAPE, 
                                           mode='constant', preserve_range=True)
            
            # convert image from H*W*C to C*H*w
            img_temp = img_temp.transpose((2, 0, 1))
            for channel in range(img_temp.shape[0]):
                img_temp[channel, :, :] = (img_temp[channel, :, :] - ave[channel])/std[channel]
                
            # for mask
            r = cv2.resize(mask_temp[:, :, 0], (config.IMG_SHAPE[1],config.IMG_SHAPE[0]), interpolation=cv2.INTER_NEAREST)
            g = cv2.resize(mask_temp[:, :, 1], (config.IMG_SHAPE[1],config.IMG_SHAPE[0]), interpolation=cv2.INTER_NEAREST)
            b = cv2.resize(mask_temp[:, :, 2], (config.IMG_SHAPE[1],config.IMG_SHAPE[0]), interpolation=cv2.INTER_NEAREST)
            
            label_temp = np.zeros(config.IMG_SHAPE, dtype=np.float32)
            non_healing_index = (r == colorKeyGT['Non-healing'][0]) & (g == colorKeyGT['Non-healing'][1]) & (b == colorKeyGT['Non-healing'][2])
            label_temp[non_healing_index] = 1
            
            # morphological operation
            label_temp = skimage.morphology.closing(label_temp,disk(3))
            
            
            
            # Add unknown category
            # index1 = r == COLORKEY_GT['Unknown_Category'][0]
            # index2 = g == COLORKEY_GT['Unknown_Category'][1]
            # index3 = b == COLORKEY_GT['Unknown_Category'][2]
            # label_temp[(index1 & index2 & index3)] = 255
            
            
            # Augmentation
#             if augmentation:
#                 img_temp, label_temp = augmentor(img_temp, label_temp)
                
            img_batch[index, ...] = img_temp
            label_batch[index, ...] = label_temp
        
        
#         print('img_batch: {}, label_batch:{}'.format(np.sum(img_batch),np.unique(label_batch)))
        yield img_batch, label_batch
```

## 3) SegNet Model

```python
from __future__ import print_function,division
import torch
import torch.nn as nn
import torch.nn.functional as F

class SegNet(nn.Module):
    """
    segnet algorithm
    """
    def __init__(self,inputs,labels):
        super(SegNet,self).__init__()
        
        batchNorm_momentum = 0.1
        self.inputs = inputs
        
        # block 1
        self.conv11 = nn.Conv2d(inputs,64,kernel_size=3,padding=1)
        self.bn11 = nn.BatchNorm2d(64,momentum=batchNorm_momentum)
        self.conv12 = nn.Conv2d(64,64,kernel_size=3,padding=1)
        self.bn12 = nn.BatchNorm2d(64,momentum=batchNorm_momentum)
        
        # block2
        self.conv21 = nn.Conv2d(64,128,kernel_size=3,padding=1)
        self.bn21 = nn.BatchNorm2d(128,momentum=batchNorm_momentum)
        self.conv22 = nn.Conv2d(128,128,kernel_size=3,padding=1)
        self.bn22 = nn.BatchNorm2d(128,momentum=batchNorm_momentum)
        
        #block3
        self.conv31 = nn.Conv2d(128,256,kernel_size=3,padding=1)
        self.bn31 = nn.BatchNorm2d(256,momentum=batchNorm_momentum)
        self.conv32 = nn.Conv2d(256,256,kernel_size=3,padding=1)
        self.bn32 = nn.BatchNorm2d(256,momentum=batchNorm_momentum)
        self.conv33 = nn.Conv2d(256,256,kernel_size=3,padding=1)
        self.bn33 = nn.BatchNorm2d(256,momentum=batchNorm_momentum)
        
        #block4
        self.conv41 = nn.Conv2d(256,512,kernel_size=3,padding=1)
        self.bn41 = nn.BatchNorm2d(512,momentum=batchNorm_momentum)
        self.conv42 = nn.Conv2d(512,512,kernel_size=3,padding=1)
        self.bn42 = nn.BatchNorm2d(512,momentum=batchNorm_momentum)
        self.conv43 = nn.Conv2d(512,512,kernel_size=3,padding=1)
        self.bn43 = nn.BatchNorm2d(512,momentum=batchNorm_momentum)
        
        #block5
        self.conv51 = nn.Conv2d(512,512,kernel_size=3,padding=1)
        self.bn51 = nn.BatchNorm2d(512,momentum=batchNorm_momentum)
        self.conv52 = nn.Conv2d(512,512,kernel_size=3,padding=1)
        self.bn52 = nn.BatchNorm2d(512,momentum=batchNorm_momentum)
        self.conv53 = nn.Conv2d(512,512,kernel_size=3,padding=1)
        self.bn53 = nn.BatchNorm2d(512,momentum=batchNorm_momentum)
        
        # decode 5
        self.conv53d = nn.Conv2d(512,512,kernel_size=3,padding=1)
        self.bn53d = nn.BatchNorm2d(512,momentum=batchNorm_momentum)
        self.conv52d = nn.Conv2d(512, 512, kernel_size=3, padding=1)
        self.bn52d = nn.BatchNorm2d(512, momentum=batchNorm_momentum)
        self.conv51d = nn.Conv2d(512, 512, kernel_size=3, padding=1)
        self.bn51d = nn.BatchNorm2d(512, momentum=batchNorm_momentum)
        
        # decode 4
        self.conv43d = nn.Conv2d(512, 512, kernel_size=3, padding=1)
        self.bn43d = nn.BatchNorm2d(512, momentum=batchNorm_momentum)
        self.conv42d = nn.Conv2d(512, 512, kernel_size=3, padding=1)
        self.bn42d = nn.BatchNorm2d(512, momentum=batchNorm_momentum)
        self.conv41d = nn.Conv2d(512, 256, kernel_size=3, padding=1)
        self.bn41d = nn.BatchNorm2d(256, momentum=batchNorm_momentum)
        
        # decode 3
        self.conv33d = nn.Conv2d(256, 256, kernel_size=3, padding=1)
        self.bn33d = nn.BatchNorm2d(256, momentum=batchNorm_momentum)
        self.conv32d = nn.Conv2d(256, 256, kernel_size=3, padding=1)
        self.bn32d = nn.BatchNorm2d(256, momentum=batchNorm_momentum)
        self.conv31d = nn.Conv2d(256,  128, kernel_size=3, padding=1)
        self.bn31d = nn.BatchNorm2d(128, momentum=batchNorm_momentum)
        
        # decode 2
        self.conv22d = nn.Conv2d(128, 128, kernel_size=3, padding=1)
        self.bn22d = nn.BatchNorm2d(128, momentum=batchNorm_momentum)
        self.conv21d = nn.Conv2d(128, 64, kernel_size=3, padding=1)
        self.bn21d = nn.BatchNorm2d(64, momentum=batchNorm_momentum)
        
        # decode 1
        self.conv12d = nn.Conv2d(64, 64, kernel_size=3, padding=1)
        self.bn12d = nn.BatchNorm2d(64, momentum=batchNorm_momentum)
        self.conv11d = nn.Conv2d(64, labels, kernel_size=3, padding=1)
        
    
    def forward(self,x):
        """
        forward the network
        """
        # encode 1
        x11 = F.relu(self.bn11(self.conv11(x)))
        x12 = F.relu(self.bn12(self.conv12(x11)))
        x1p,id1 = F.max_pool2d(x12,kernel_size=2,stride=2,return_indices=True)
        size1 = x12.size()
        
        # encode2
        x21 = F.relu(self.bn21(self.conv21(x1p)))
        x22 = F.relu(self.bn22(self.conv22(x21)))
        x2p,id2 = F.max_pool2d(x22,kernel_size=2,stride=2,return_indices=True)
        size2 = x22.size()
        
        # encode3
        x31 = F.relu(self.bn31(self.conv31(x2p)))
        x32 = F.relu(self.bn32(self.conv32(x31)))
        x33 = F.relu(self.bn33(self.conv33(x32)))
        x3p,id3 = F.max_pool2d(x33,kernel_size=2,stride=2,return_indices=True)
        size3 = x33.size()
        
        # encode4
        x41 = F.relu(self.bn41(self.conv41(x3p)))
        x42 = F.relu(self.bn42(self.conv42(x41)))
        x43 = F.relu(self.bn43(self.conv43(x42)))
        x4p,id4 = F.max_pool2d(x43,kernel_size=2,stride=2,return_indices=True)
        size4 = x43.size()
        
        # encode5
        x51 = F.relu(self.bn51(self.conv51(x4p)))
        x52 = F.relu(self.bn52(self.conv52(x51)))
        x53 = F.relu(self.bn53(self.conv53(x52)))
        x5p,id5 = F.max_pool2d(x53,kernel_size=2,stride=2,return_indices=True)
        size5 = x53.size()
        
        #==========================================>>>>>>>>>>>
        
        # decode 5
        x5d = F.max_unpool2d(x5p,id5,kernel_size=2,stride=2,output_size=size5)
        x53d = F.relu(self.bn53d(self.conv53d(x5d)))
        x52d = F.relu(self.bn52d(self.conv52d(x53d)))
        x51d = F.relu(self.bn51d(self.conv51d(x52d)))
        
        # decode4
        x4d = F.max_unpool2d(x51d, id4, kernel_size=2, stride=2, output_size=size4)
        x43d = F.relu(self.bn43d(self.conv43d(x4d)))
        x42d = F.relu(self.bn42d(self.conv42d(x43d)))
        x41d = F.relu(self.bn41d(self.conv41d(x42d)))
        
        # decode3
        x3d = F.max_unpool2d(x41d, id3, kernel_size=2, stride=2, output_size=size3)
        x33d = F.relu(self.bn33d(self.conv33d(x3d)))
        x32d = F.relu(self.bn32d(self.conv32d(x33d)))
        x31d = F.relu(self.bn31d(self.conv31d(x32d)))
        
        # decode2
        x2d = F.max_unpool2d(x31d, id2, kernel_size=2, stride=2, output_size=size2)
        x22d = F.relu(self.bn22d(self.conv22d(x2d)))
        x21d = F.relu(self.bn21d(self.conv21d(x22d)))
        
        # decode1
        x1d = F.max_unpool2d(x21d, id1, kernel_size=2, stride=2, output_size=size1)
        x12d = F.relu(self.bn12d(self.conv12d(x1d)))
        x11d = self.conv11d(x12d)
        
        return x11d
    
    def initialized_with_pretrained_weights(self):
        """Initialiaze."""
        corresp_name = {
                "features.0.weight": "conv11.weight",
                "features.0.bias": "conv11.bias",
                "features.1.weight": "bn11.weight",
                "features.1.bias": "bn11.bias",
                "features.1.running_mean": "bn11.running_mean",
                "features.1.running_var": "bn11.running_var",

                "features.3.weight": "conv12.weight",
                "features.3.bias": "conv12.bias",
                "features.4.weight": "bn12.weight",
                "features.4.bias": "bn12.bias",
                "features.4.running_mean": "bn12.running_mean",
                "features.4.running_var": "bn12.running_var",

                "features.7.weight": "conv21.weight",
                "features.7.bias": "conv21.bias",
                "features.8.weight": "bn21.weight",
                "features.8.bias": "bn21.bias",
                "features.8.running_mean": "bn21.running_mean",
                "features.8.running_var": "bn21.running_var",

                "features.10.weight": "conv22.weight",
                "features.10.bias": "conv22.bias",
                "features.11.weight": "bn22.weight",
                "features.11.bias": "bn22.bias",
                "features.11.running_mean": "bn22.running_mean",
                "features.11.running_var": "bn22.running_var",

                # stage 3
                "features.14.weight": "conv31.weight",
                "features.14.bias": "conv31.bias",
                "features.15.weight": "bn31.weight",
                "features.15.bias": "bn31.bias",
                "features.15.running_mean": "bn31.running_mean",
                "features.15.running_var": "bn31.running_var",

                "features.17.weight": "conv32.weight",
                "features.17.bias": "conv32.bias",
                "features.18.weight": "bn32.weight",
                "features.18.bias": "bn32.bias",
                "features.18.running_mean": "bn32.running_mean",
                "features.18.running_var": "bn32.running_var",

                "features.20.weight": "conv33.weight",
                "features.20.bias": "conv33.bias",
                "features.21.weight": "bn33.weight",
                "features.21.bias": "bn33.bias",
                "features.21.running_mean": "bn33.running_mean",
                "features.21.running_var": "bn33.running_var",

                # stage 4
                "features.24.weight": "conv41.weight",
                "features.24.bias": "conv41.bias",
                "features.25.weight": "bn41.weight",
                "features.25.bias": "bn41.bias",
                "features.25.running_mean": "bn41.running_mean",
                "features.25.running_var": "bn41.running_var",

                "features.27.weight": "conv42.weight",
                "features.27.bias": "conv42.bias",
                "features.28.weight": "bn42.weight",
                "features.28.bias": "bn42.bias",
                "features.28.running_mean": "bn42.running_mean",
                "features.28.running_var": "bn42.running_var",

                "features.30.weight": "conv43.weight",
                "features.30.bias": "conv43.bias",
                "features.31.weight": "bn43.weight",
                "features.31.bias": "bn43.bias",
                "features.31.running_mean": "bn43.running_mean",
                "features.31.running_var": "bn43.running_var",

                # stage 5
                "features.34.weight": "conv51.weight",
                "features.34.bias": "conv51.bias",
                "features.35.weight": "bn51.weight",
                "features.35.bias": "bn51.bias",
                "features.35.running_mean": "bn51.running_mean",
                "features.35.running_var": "bn51.running_var",

                "features.37.weight": "conv52.weight",
                "features.37.bias": "conv52.bias",
                "features.38.weight": "bn52.weight",
                "features.38.bias": "bn52.bias",
                "features.38.running_mean": "bn52.running_mean",
                "features.38.running_var": "bn52.running_var",

                "features.40.weight": "conv53.weight",
                "features.40.bias": "conv53.bias",
                "features.41.weight": "bn53.weight",
                "features.41.bias": "bn53.bias",
                "features.41.running_mean": "bn53.running_mean",
                "features.41.running_var": "bn53.running_var",
                }
        # load the state dict of pretrained model
        import torch.utils.model_zoo as model_zoo
        pretrained_sd = model_zoo.load_url("http://download.pytorch.org/models/vgg16_bn-6c64b313.pth")
        s_dict = self.state_dict()
        for name in pretrained_sd:
            if name not in corresp_name:
                continue
            if("features.0" not in name) or (self.inputs==3):
                s_dict[corresp_name[name]] = pretrained_sd[name]
        self.load_state_dict(s_dict)
        
    
    def load_from_filename(self,model_path):
        """Load weights method."""
        th = torch.load(model_path)  # load the weigths
        self.load_state_dict(th)
        
    def load_from_filename_revise(self,model_path):
        pretrained_sd = torch.load(model_path)  # load the weights
        s_dict = self.state_dict()
        for name in pretrained_sd:
            if("11" not in name):
                s_dict[name] = pretrained_sd[name]
   #             print('shape-0:{}'.format(s_dict[name].shape))
        self.load_state_dict(s_dict)
        
    def load_from_dict(self,caffe_dict):
        s_dict = self.state_dict()
        
        for name in s_dict:
            if name in caffe_dict:
                s_dict[name] = caffe_dict[name]
             #   print('shape-1:{}'.format(s_dict[name].shape))
                
        self.load_state_dict(s_dict)
```

```python
import torch.nn.init as init

def weights_init(m):
    if isinstance(m, nn.Conv2d):
        init.kaiming_normal_(m.weight.data)
        m.bias.data.fill_(0.01)
    elif isinstance(m,nn.BatchNorm2d):
        m.weight.data.fill_(1)
        m.bias.data.fill_(0.01)
```

### save caffe weights as dict used in pytorch

```python
import numpy as np
import os

def get_caffe_dict(pytorch_net,caffe_params):
    corresp_name = {
        "11":"conv1_1_1",
        "12":"conv1_2",
        "21":"conv2_1",
        "22":"conv2_2",
        "31":"conv3_1",
        "32":"conv3_2",
        "33":"conv3_3",
        "41":"conv4_1",
        "42":"conv4_2",
        "43":"conv4_3",
        "51":"conv5_1",
        "52":"conv5_2",
        "53":"conv5_3",
        "53d":"conv5_3_D",
        "52d":"conv5_2_D",
        "51d":"conv5_1_D",
        "43d":"conv4_3_D",
        "42d":"conv4_2_D",
        "41d":"conv4_1_D",
        "33d":"conv3_3_D",
        "32d":"conv3_2_D",
        "31d":"conv3_1_D",
        "22d":"conv2_2_D",
        "21d":"conv2_1_D",
        "12d":"conv1_2_D",
        "11d":"conv1_1_1_D",
    }
    
    final_dict = {}
    for name in pytorch_net.state_dict():
        base_name = name.split('.')[0]
        
        if 'd' in base_name:
            base_name_2 = base_name[-3:]
        else:
            base_name_2 = base_name[-2:]
            
        if '11' in base_name_2 and 'd' not in base_name_2:
            continue
            
        base_name_caffe = corresp_name[base_name_2]
#         print(base_name_caffe)
            
        if ('conv' in name) and ('weight' in name):
            temp = np.load(caffe_params + os.sep + base_name_caffe + '_0.npy')
            if np.all(temp==0):
                continue
            else:
                final_dict[name] = torch.from_numpy(temp)

        elif ('conv' in name) and ('bias' in name):
            temp = np.load(caffe_params + os.sep + base_name_caffe + '_1.npy')
            if np.all(temp==0):
                continue
            else:
                final_dict[name] = torch.from_numpy(temp)

        elif ('bn' in name) and ('running_mean' in name):
            temp = np.load(caffe_params + os.sep + base_name_caffe + '_bn_tmp_0.npy')
            if np.all(temp==0):
                continue
            else:
                final_dict[name] = torch.from_numpy(temp)

        elif ('bn' in name) and ('running_var' in name):
            temp = np.load(caffe_params + os.sep + base_name_caffe + '_bn_tmp_1.npy')
            if np.all(temp==0):
                continue
            else:
                final_dict[name] = torch.from_numpy(temp)

        elif ('bn' in name) and ('weight' in name):
            temp = np.load(caffe_params + os.sep + base_name_caffe + '_scale_0.npy')
            if np.all(temp==0):
                continue
            else:
                final_dict[name] = torch.from_numpy(temp)

        elif ('bn' in name) and ('bias' in name):
            temp = np.load(caffe_params + os.sep + base_name_caffe + '_scale_1.npy')
            if np.all(temp==0):
                continue
            else:
                final_dict[name] = torch.from_numpy(temp)

        else:
            print("Error: No 'conv' and 'bn'")  
                
    return final_dict
```

## 4) Algorithm Training
### 4.1: define training function 4.2: leave one out cross validation 4.3: inference 4.4: show the rpediction results

```python
import torch.backends.cudnn as cudnn
from torch import optim
from torch.autograd import Variable
from torch.optim.lr_scheduler import MultiStepLR

def SegNet_train(net,config,avg,std,test_index=0,shuffle=True):
#     print("""
#     Starting training:
#     Epochs:{}
#     Batch Size: {}
#     Learning rate: {}
#     Which GPU: {}
#     """.format(config.NUM_EPOCHES,config.BATCH_SIZE,config.BASE_LR,config.WHICH_GPU))
    
    # Set CUDA mode
    if torch.cuda.is_available() & config.GPU_COUNT:
        device = torch.device("cuda", index=config.WHICH_GPU)
    else:
        print("Using CPU to train the model!")
        device = torch.device("cpu")
            
            
    caffe_params = config.WEIGHTS_IN_CAFFE
    net.apply(weights_init)
    final_dict_from_caffe = get_caffe_dict(net,caffe_params)
    
    net.to(device)
    
#     net.initialized_with_pretrained_weights()  # using VGG to initialize the network
    net.load_from_dict(final_dict_from_caffe)
    
    optimizer = optim.SGD(net.parameters(),lr=config.BASE_LR,momentum=config.MOMENTUM,weight_decay=config.WEIGH_DECAY)
    
    class_weights = torch.FloatTensor([0.9,5.0]).to(device)   
    criterion = nn.CrossEntropyLoss(weight=class_weights,ignore_index=255)  # combined with net = UNet(8,2), handle ignore pixel
    
    
    loss_history = []
    accuracy_history = []
    for epoch in range(config.NUM_EPOCHES):
        img_generator = generator(config, 
                                  avg, 
                                  std, 
                                  indicator = "train", 
                                  index=test_index, 
                                  batch_size=config.BATCH_SIZE, 
                                  shuffle=shuffle, 
                                  augmentation=True)
        
        epoch_loss = []
        epoch_accuracy = []
        
         # Each step
        for inputs,labels in img_generator:
            # NumPy to PyTorch
            inputs = torch.from_numpy(inputs)
            labels = torch.from_numpy(labels).long()

            # Put to device
            inputs = inputs.to(device)
            labels = labels.to(device)  #[batch_size, height, width]
            
            net.train()
            
            # Forward
            outputs = net(inputs)  #[batch_size, n_classes, height, width]
            
            loss = criterion(outputs,labels)
            
            epoch_loss.append(loss.item())
            loss_history.append(loss.item())
            
            
            # Accuracy
            accuracy = []
            pred_labels = F.softmax(outputs).cpu().detach().numpy()
            pred_labels = np.argmax(pred_labels, axis=1).astype('int')  # [batch_size, h, w]
            true_labels = labels.cpu().numpy()
            
            # Total accuracy
            accuracy.append(float(np.sum(true_labels == pred_labels)) / (config.IMG_SHAPE[0] * config.IMG_SHAPE[1] * config.BATCH_SIZE))
            for n_label in range(config.NUM_CLASSES):
                #Accuracy for each class
                n_target = np.sum(true_labels == n_label)
                if n_target > 0:
                    accuracy.append(float(np.sum((true_labels == n_label) & (pred_labels == n_label))) / n_target)
                else:
                    accuracy.append(np.nan)
            accuracy_history.append(accuracy)
            epoch_accuracy.append(accuracy)
            
            
            # Backward
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        
        # Record loss for each epoch
        print("Epoch {}, loss: {}, accuracy: {}".format(epoch + 1, np.average(epoch_loss), np.average(epoch_accuracy, axis=0)))    
          
```

```python
def start_train(config):
    """
    leave one out cross validation
    """
    # calculate average and std
    avg,std = get_avg_std_info(config)
    
    # total number of DFU
    os.chdir(config.IMG_FOLDER)
    imgs = glob("*day*.png")    
    num_total = len(imgs)
    
    for index in range(num_total):
        print('\nTraining --- Processing folder {}/{}'.format(index+1,num_total))
        net = SegNet(3,2)
        SegNet_train(net,config,avg,std,test_index=index,shuffle=True)
        
        # save the model
        saveModelFolder = config.MODEL_DIR + os.sep + 'folder' + str(index + 1)
       

        if not os.path.exists(saveModelFolder):
            os.makedirs(saveModelFolder)
       
        # saving the model
        torch.save(net.state_dict(),saveModelFolder + os.sep + 'segNet.pth')
```

```python
from sklearn.metrics import roc_auc_score
import pandas_ml as pdml

def inference(net,config,avg,std,index=0,toplot=True):
    """
    do inference
    """
    
    # Set CUDA mode
    if torch.cuda.is_available() & config.GPU_COUNT:
        device = torch.device("cuda", index=config.WHICH_GPU)
    else:
        print("Using CPU to train the model!")
        device = torch.device("cpu")
        
    model_file = config.MODEL_DIR + os.sep + 'folder' + str(index + 1) + os.sep + 'segNet.pth'
        
    print("Loading model from {}...".format(model_file))
    net.to(device)
    net.load_state_dict(torch.load(model_file))
    
    
    savePredFolder = config.RESULTS_DIR + os.sep + 'folder' + str(index + 1)
    if not os.path.exists(savePredFolder):
        os.makedirs(savePredFolder)
        
    img_generator = generator(config,
                              avg,
                              std, 
                              indicator = "test",
                              index=index,
                              batch_size=1,
                              augmentation=False)
    
    for i,(inputs,labels) in enumerate(img_generator):
        # NumPy to PyTorch
        inputs = torch.from_numpy(inputs)
#         labels = torch.from_numpy(labels)

        # Put to device
        inputs = inputs.to(device)
        labels = labels[0].astype('uint8')  #[batch_size, height, width]
        
        net.eval()
        with torch.no_grad():
            output = net(inputs)
#             print(output.shape)
            output = F.softmax(output.squeeze()).cpu().numpy()
            pred_img = np.argmax(output, axis=0).astype('uint8')
            non_healing_prob = output[1, :, :]
            
        masked = np.ma.masked_where(pred_img == 0, pred_img)

        # Save images, labels, and prediction
        scipy.misc.imsave(os.path.join(savePredFolder, 'prediction_{}.tif'.format(i)), 255 * pred_img.astype(float))
        scipy.misc.toimage(255*labels.astype(float), cmin=0.0, cmax=255.0).save(os.path.join(savePredFolder, 
                                                                                       'mask_{}.tif'.format(i)))
        
        # plot image
        if toplot:
            plt.figure(figsize=(12, 4))
            plt.subplot(151)
            image = np.transpose(inputs[0], (1, 2, 0))
            for ch in range(config.NUM_CHANNEL):
                image[:, :, ch] = (image[:, :, ch] * std[ch]) + avg[ch]
            image = np.array(image, dtype=np.uint8)
            plt.imshow(image,aspect="auto")
            plt.axis('off')
            plt.title("Original Image")
            plt.subplot(152)
            plt.imshow(non_healing_prob,aspect="auto")
            plt.axis('off')
            plt.title("Prob")
            plt.colorbar()
            plt.subplot(153)
            plt.imshow(pred_img,aspect="auto")
            plt.axis('off')
            plt.title("Predicted Mask")
            plt.subplot(154)
            plt.imshow(image, 'gray', interpolation='none',aspect="auto")
            plt.imshow(masked, 'jet', interpolation='none', alpha=0.7,aspect="auto")
            plt.axis('off')
            plt.title("Predicted mask on image")
            plt.show()
            plt.subplot(155)
            plt.imshow(labels,aspect="auto")
            plt.axis('off')
            plt.title("Ground Truth")
            plt.show()
            
            
        # calculate the sensitivity, specificity, AUC, accuracy from labels,pred_img,non_healing_prob

        
#         pdml_cm = pdml.ConfusionMatrix(labels.ravel(),pred_img.ravel())
#         stats = pdml_cm.stats()

        
        sens = (1.0 * np.sum((labels == pred_img) & (pred_img == 1))) / np.sum(labels == 1)
        spec = (1.0 * np.sum((labels == pred_img) & (pred_img == 0))) / np.sum(labels == 0)
        
#         print('sens:{}/{}'.format(sens,stats['TPR']))
#         print('spec:{}/{}'.format(spec,stats['TNR']))
        acc = 1.0 * np.sum(labels == pred_img) / (config.IMG_SHAPE[0] * config.IMG_SHAPE[1])
        auc = roc_auc_score(labels.ravel(),non_healing_prob.ravel())
        
        return sens,spec,acc,auc
                                                   
```

```python
def training_inference(net,config,avg,std,index=0,toplot=True):
    """
    do inference
    """
    
    # Set CUDA mode
    if torch.cuda.is_available() & config.GPU_COUNT:
        device = torch.device("cuda", index=config.WHICH_GPU)
    else:
        print("Using CPU to train the model!")
        device = torch.device("cpu")
        
    model_file = config.MODEL_DIR + os.sep + 'folder' + str(index + 1) + os.sep + 'segNet.pth'
        
    print("Loading model from {}...".format(model_file))
    net.to(device)
    net.load_state_dict(torch.load(model_file))
        
    img_generator = generator(config,
                              avg,
                              std, 
                              indicator = "train",
                              index=index,
                              batch_size=1,
                              augmentation=False)
    
    for i,(inputs,labels) in enumerate(img_generator):
        # NumPy to PyTorch
        inputs = torch.from_numpy(inputs)

        # Put to device
        inputs = inputs.to(device)
        labels = labels[0].astype('uint8')  #[batch_size, height, width]
        
        net.eval()
        with torch.no_grad():
            output = net(inputs)
            output = F.softmax(output.squeeze()).cpu().numpy()
            pred_img = np.argmax(output, axis=0).astype('uint8')
            non_healing_prob = output[1, :, :]
        
        masked = np.ma.masked_where(pred_img == 0, pred_img)
        # plot image
        if toplot:
            plt.figure(figsize=(16, 5))
            plt.subplot(151)
            image = np.transpose(inputs[0], (1, 2, 0))
            for ch in range(config.NUM_CHANNEL):
                image[:, :, ch] = (image[:, :, ch] * std[ch]) + avg[ch]
            image = np.array(image, dtype=np.uint8)
            plt.imshow(image,aspect="auto")
            plt.title("Original Image")
            plt.subplot(152)
            plt.imshow(non_healing_prob,aspect="auto")
            plt.title("Prob")
            plt.colorbar()
            plt.subplot(153)
            plt.imshow(pred_img,aspect="auto")
            plt.title("Predicted Mask")
            plt.subplot(154)
            plt.imshow(labels,aspect="auto")
            plt.title("Ground Truth")
            plt.show()
            plt.subplot(155)
            plt.imshow(image, 'gray', interpolation='none',aspect="auto")
            plt.imshow(masked, 'jet', interpolation='none', alpha=0.5,aspect="auto")
            
        # calculate the sensitivity, specificity, AUC, accuracy from labels,pred_img,non_healing_prob
        sens = (1.0 * np.sum((labels == pred_img) & (pred_img == 1))) / np.sum(labels == 1)
        spec = (1.0 * np.sum((labels == pred_img) & (pred_img == 0))) / np.sum(labels == 0)
        acc = 1.0 * np.sum(labels == pred_img) / (config.IMG_SHAPE[0] * config.IMG_SHAPE[1])
        auc = roc_auc_score(labels.ravel(),non_healing_prob.ravel())
        
        print('sens:{},spec:{},acc:{},auc:{}'.format(sens,spec,acc,auc))
                                                   
```
